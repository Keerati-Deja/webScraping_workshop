```python
page = 0

ls = []
brand_list = []
product_list = []
type_list = []
datetime_list = []
review_list = []
rate_list = []

emp_ls = []
user_ls = []
skin_ls = []
age_range_ls = []
rev_count_ls = []

ben_ls = []
shop_ls = []
when_ls = []

while page <= 10:
    data = requests.get('https://www.vanilla.in.th/allreview.cgi?page='+str(page))
    soup = bs4.BeautifulSoup(data.content)
    for item in soup.find_all('div', {'class':'all-reviews'}):
        #brands & products
        ls.append(item.find('div',{'class':'header6'}).text)
        brand_ls, prod_ls = zip(*map(lambda x: x.split('\n'), ls))
        brand_list = list(brand_ls)
        product_list = list(prod_ls)
        #product type
        type_list.append(item.find('p',{'style':'margin-bottom:5px;'}).text.strip('-').strip())
        #reviews
        review_list.append(item.find('p',{'class':'thaiSTD comment'}).text.replace('\r','').replace('\n','').strip())
        #rates
        rate_list.append(item.find('li',{'class':'stars clearfix colRight'}).text.replace('\r','').replace('\n','').strip())
        #datetime review
        datetime_list.append(item.find('div',{'class':'date-all'}).text)

    for index, element in enumerate(soup.find_all('span', {'class':'hide-phone'})):
        if index % 5 == 0:
            emp_ls.append(element.text.strip())
        elif index % 5 == 1:
            user_ls.append(element.text.strip())
        elif index % 5 == 2:
            skin_ls.append(element.text.strip())
        elif index % 5 == 3:
            age_range_ls.append(element.text.strip())
        else:
            rev_count_ls.append(element.text.strip())
                
    for index, element in enumerate(soup.find_all('li',{'class','clearfix colRight'})):
        if index % 3 == 0:
            ben_ls.append(element.text.strip())
        elif index % 3 == 1:
                shop_ls.append(element.text.strip())
        else:
            when_ls.append(element.text.strip())

    print(f"Completed page {page+1}")
    page += 1
```

```python
table1 = pd.DataFrame([user_ls,skin_ls,age_range_ls,rev_count_ls,ben_ls,shop_ls,when_ls]).transpose()
table1.columns = ['username','skin_type','age_range','rev_count','benefits','shop_at','rev_when']
table1
```
