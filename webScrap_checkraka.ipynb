Scraping many pages

```python
# Build an automatic Bot

import requests
import bs4
import pandas as pd

# Strating at page 1
page = 1

name_list = []
price_list = []

# until the last page (number = 17, in this case)
while page <= 17:
    # increse the number of page, convert int to str for the link format
    data_ = requests.get('https://www.checkraka.com/car/search?perPage=60&page='+str(page))
    soup_ = bs4.BeautifulSoup(data_.text)
    for car_name in soup_.find_all('div',{'class':'product-highlight-card'}):
        name_list.append(car_name.find('div',{'class':'card'}).find('span').text)
        price_list.append(car_name.find('div',{'class':'card'}).find('label').text.split()[0].replace(',',''))
    print('Complete page number: ', page)
    page += 1
    
table = pd.DataFrame([name_list,price_list]).transpose()
table.columns = ['name','price']
table.set_index('name')
```

```python
# Save the complete dataframe into csv file
table.to_csv('car-price_checkraka.csv', index=False)
```
